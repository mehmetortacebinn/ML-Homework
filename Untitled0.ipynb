```python
#3.	Veri setinizi rastgele olarak %70 eitim %30 test olacak ekilde ayrnz. Eitim veri seti iin oklu Dorusal Regresyon analizi ve Multinominal Lojistik Regresyon analizi yntemlerini uygulaynz.
#Elde ettiiniz katsaylar raporlaynz.  Elde ettiiniz regresyon denklemlerini kullanarak Test kmesi iin performans metriklerini hesaplaynz.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_score, classification_report

# Veriyi eitim ve test setlerine ayr
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

pca = PCA(n_components=2)
X_train = pca.fit_transform(X_train, y_train)
X_test = pca.transform(X_test)


# oklu Dorusal Regresyon modelini eit
linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)

# Multinominal Lojistik Regresyon modelini eit
logistic_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs')
log = logistic_reg.fit(X_train, Y_train)

# Katsaylar raporla
print("oklu Dorusal Regresyon Katsaylar:")
print(linear_reg.coef_)
print("\nMultinominal Lojistik Regresyon Katsaylar:")
print(logistic_reg.coef_)

# oklu Dorusal Regresyon denklemi
linear_eq = f"y = {linear_reg.intercept_} "
for i, coef in enumerate(linear_reg.coef_):
    linear_eq += f"+ ({coef} * X{i})"

print("\noklu Dorusal Regresyon Denklemi:")
print(linear_eq)

# Multinominal Lojistik Regresyon denklemi
logistic_eq = "y = "
for i, coef in enumerate(logistic_reg.coef_[0]):
    logistic_eq += f"({coef} * X{i}) "
print("\nMultinominal Lojistik Regresyon Denklemi:")
print(logistic_eq)

# Test kmesi performans metrikleri
linear_reg_predictions = linear_reg.predict(X_test)
logistic_reg_predictions = logistic_reg.predict(X_test)

linear_reg_mse = mean_squared_error(Y_test, linear_reg_predictions)
logistic_reg_accuracy = accuracy_score(Y_test, logistic_reg_predictions)

print("\noklu Dorusal Regresyon MSE:", linear_reg_mse)
print("Multinominal Lojistik Regresyon Doruluu:", logistic_reg_accuracy)

Y_pred = log.predict(X_test)

cm = confusion_matrix(Y_test, Y_pred)

print("Karklk Matrisi:")
print(cm)

# Karklk matrisini grselletir
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g',
            xticklabels=['Tahmin 0', 'Tahmin 1'],
            yticklabels=['Gerek 0', 'Gerek 1'])
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerek Deer')
plt.title('Karklk Matrisi')
plt.show()

# Performans metriklerini hesapla
accuracy = accuracy_score(Y_test, Y_pred)
classification_rep = classification_report(Y_test, Y_pred)

print("\nTest Verisi in Performans Metrikleri:")
print("Doruluk:", accuracy)
print("Snflandrma Raporu:\n", classification_rep)

```
